```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger
```

```python
patterns = [
    (r'.*ing$', 'VBG'),        # gerunds (running, eating)
    (r'.*ed$', 'VBD'),         # past tense verbs
    (r'.*es$', 'VBZ'),         # 3rd person singular verbs
    (r'.*ould$', 'MD'),        # modals (would, could)
    (r'.*\'s$', 'NN$'),        # possessive nouns
    (r'.*s$', 'NNS'),          # plural nouns
    (r'^-?[0-9]+(\.[0-9]+)?$', 'CD'),  # cardinal numbers
    (r'.*', 'NN')              # default: noun
]
```

```python
regex_tagger = RegexpTagger(patterns)
```

```python
sentence = "The children were playing games and eating snacks"
tokens = word_tokenize(sentence)

tagged = regex_tagger.tag(tokens)
print(tagged)
```
